현재 데이터는 연도별로 따로 있다. 하나의 데이터로 통합하기 위해서 아래와 같은 작업을 거친다.

```python
targets = ['Low', 'Low-Mid', 'Top-Mid', 'Top']
h_cols = ['Country', 'GDP', 'Family', 'Life', 'Freedom', 'Generosity', 'Trust']
```

```python
def prep_frame(df_year, year):
    df = pd.DataFrame()
    
    # 2015, 2016, 2017 데이터를 통합하는 작업
    target_cols = []
    for c in h_cols:
        # h_cols의 값이 x에 있다면 target_cols에 추가
        target_cols.extend([x for x in df_year.columns if c in x]) 
    
    # df_year에서 해당 columns의 값들을 df로 복사
    df[h_cols] = df_year[target_cols]
    
    print(df)
    
    # Score라는 단어가 있으면 df의 'Happiness Score' column에 복사
    df['Happiness Score'] = df_year[[x for x in df_year.columns if 'Score' in x]]
    
    print(df)
    
    # 데이터 quartile값에 따라서 레이블 숫자를 배정
    # pd.qcut() 함수 사용
    # https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.qcut.html
    # 문자형 target 변수
    # df.columns[-1]은 df에서 가장 마지막 column
    df["target"] = pd.qcut(df[df.columns[-1]], len(targets), labels=targets) 
    
    print(df)
    
    # 숫자형 target 변수
    # df.colums[-2]는 df에서 마지막 두번째 column
    df["target_n"] = pd.qcut(df[df.columns[-2]], len(targets), labels=range(len(targets))) 
    
    print(df)
    
    # 연도를 추가하고, ['Country', 'Year']를 인덱스로 설정함
    df['Year'] = year
    df = df.set_index(['Country', 'Year'])
    
    return df
```

```python
df = prep_frame(df_2015, 2015) # df_2015 데이터에 대한 정리 완료
print(df)
```

```python
df
```

```python
df = df.append(prep_frame(df_2016, 2016), sort=False) # df_2016 정리된 데이터를 df_2015 다음에 붙임
df = df.append(prep_frame(df_2017, 2017), sort=False) # df_2017 정리된 데이터를 그 다음에 붙임

# 정리된 결과를 확인
df.head()
```

# 2. 모델링

## 2.1 Chapter 5 (지도학습 회귀) 실습

### 2.1.1 Define X and y

우선 X와 y를 정의한다.

```python
df
```

```python
# column이름을 사용하여 X와 y를 정의
X = df[['GDP', 'Family', 'Life', 'Freedom', 'Generosity', 'Trust']]
y_happiness = df[['Happiness Score','target_n']]
```

### 2.1.2 Training and test dataset

```python
from sklearn.model_selection import train_test_split

# 데이터 training, test 분리
X_train, X_test, y_train, y_test = train_test_split(X, y_happiness, test_size=0.2, random_state=123)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

### 2.1.3 Pre-processing

데이터 정규화를 진행한다.

```python
from sklearn.preprocessing import StandardScaler

standardScaler = StandardScaler()
standardScaler.fit(X)

X_scaled = standardScaler.transform(X)
X_train_scaled = standardScaler.transform(X_train)
X_test_scaled = standardScaler.transform(X_test)
```

### 2.1.4 Regression

회귀 분석을 진행한다.

```python
from sklearn.model_selection import cross_val_score, KFold
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

print("*****Linear Regression*****\n")

# 간단한 교차검증
kfold = KFold(n_splits=5, shuffle=True)
scores = cross_val_score(LinearRegression(), X_scaled, y_happiness['Happiness Score'], cv=kfold, scoring='r2')
print('교차검증 점수 \n{}'.format(scores))
print('교차검증 점수 평균 \n{:.3f}\n'.format(scores.mean()))
        
lm = LinearRegression()
lm.fit(X_train_scaled, y_train['Happiness Score'])

print("가중치(계수, 기울기 파라미터 W):", lm.coef_)
print("편향(절편 파라미터 b):", lm.intercept_)

print("훈련세트 점수 (r^2-score): {:.2f}".format(r2_score(lm.predict(X_train_scaled), y_train['Happiness Score'])))
print("테스트세트 점수 (r^2-score): {:.2f}".format(r2_score(lm.predict(X_test_scaled), y_test['Happiness Score'])))
```

## 2.2 Chapter 6,7 (지도학습 분류) 실습

### 2.2.1 Classification

분류 분석을 진행한다.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import roc_curve, roc_auc_score

print("*****Logistic Regression*****\n")

# 간단한 교차검증
kfold = KFold(n_splits=5, shuffle=True)
scores = cross_val_score(LogisticRegression(C=10), 
                         X_scaled, y_happiness['target_n'], cv=kfold)
print('교차검증 점수 \n{}'.format(scores))
print('교차검증 점수 평균 \n{:.3f}\n'.format(scores.mean()))

# 일반적 진행
lr = LogisticRegression(C=10)
lr.fit(X_train_scaled, y_train['target_n'])

# 정확도 결과 출력
print("훈련세트 점수 (accuracy): {:.2f}".format(lr.score(X_train_scaled, y_train['target_n'])))
print("테스트세트 점수 (accuracy): {:.2f}".format(lr.score(X_test_scaled, y_test['target_n'])))

# confusion matrix 결과 출력
print("\nConfusion matrix: \n", confusion_matrix(lr.predict(X_test_scaled), y_test['target_n']))

# classification result 결과 출력
print("\nClassification report: \n", classification_report(lr.predict(X_test_scaled), y_test['target_n']))
```

```python
from sklearn import svm

print("*****SVM*****\n")

# 간단한 교차검증
kfold = KFold(n_splits=5, shuffle=True)
scores = cross_val_score(svm.SVC(kernel='linear', random_state=123), 
                         X_scaled, y_happiness['target_n'], cv=kfold)
print('교차검증 점수 \n{}'.format(scores))
print('교차검증 점수 평균 \n{:.3f}\n'.format(scores.mean()))

# 일반적 진행
svm_model = svm.SVC(kernel='linear', random_state=123)
svm_model.fit(X_train_scaled, y_train['target_n'])

# 정확도 결과 출력
print("훈련세트 점수 (accuracy): {:.2f}".format(svm_model.score(X_train_scaled, y_train['target_n'])))
print("테스트세트 점수 (accuracy): {:.2f}".format(svm_model.score(X_test_scaled, y_test['target_n'])))

# confusion matrix 결과 출력
print("\nConfusion matrix: \n", confusion_matrix(svm_model.predict(X_test_scaled), y_test['target_n']))

# classification result 결과 출력
print("\nClassification report: \n", classification_report(svm_model.predict(X_test_scaled), y_test['target_n']))
```

```python
from sklearn.ensemble import RandomForestClassifier

print("*****Random Forest*****\n")

# 간단한 교차검증
kfold = KFold(n_splits=5, shuffle=True)
scores = cross_val_score(RandomForestClassifier(n_estimators=100, random_state=123), 
                         X_scaled, y_happiness['target_n'], cv=kfold)
print('교차검증 점수 \n{}'.format(scores))
print('교차검증 점수 평균 \n{:.3f}\n'.format(scores.mean()))

# 일반적 진행
rf = RandomForestClassifier(n_estimators=100, random_state=123)
rf.fit(X_train_scaled, y_train['target_n'])

# 정확도 결과 출력
print("훈련세트 점수 (accuracy): {:.2f}".format(rf.score(X_train_scaled, y_train['target_n'])))
print("테스트세트 점수 (accuracy): {:.2f}".format(rf.score(X_test_scaled, y_test['target_n'])))

# confusion matrix 결과 출력
print("\nConfusion matrix: \n", confusion_matrix(rf.predict(X_test_scaled), y_test['target_n']))

# classification result 결과 출력
print("\nClassification report: \n", classification_report(rf.predict(X_test_scaled), y_test['target_n']))
```

## 2.3 Chapter 8 (비지도학습 군집) 실습

k-means와 hierarchical clustering을 사용한다.

### 2.3.1 k-means

```python
from sklearn.cluster import KMeans 

# k-means clustering 실행
kmeans = KMeans(n_clusters=4)
kmeans.fit(X_scaled)
```

```python
# clustering 레이블 결과
kmeans.labels_
```

```python
# 실제 y 값 array 형태로 변경.
y_array = y_happiness['target_n'].values.to_numpy()
```

```python
# clustering 결과와 실제 y 값의 비교.
count = 0
for i in range(0,len(kmeans.labels_)):
    if kmeans.labels_[i] == y_array[i]:
        count += 1
print("겹치는 비율: {:.2f}".format(count/len(kmeans.labels_)))
```

결과가 그렇게 좋지는 않지만, clustering을 통한 grouping도 생각해볼 수 있다.

### 2.3.2 Hierarchical clustering

대표적으로 쓰이는 다른 클러스트링 방법은 Hierarchical clustring 이다 (계층적 클러스트링).

```python
from sklearn.cluster import AgglomerativeClustering

h_clustering = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
h_clustering.fit_predict(X_scaled)
```

```python
# clustering 결과와 실제 y 값의 비교
count = 0
for i in range(0,len(h_clustering.labels_)):
    if h_clustering.labels_[i] == y_array[i]:
        count += 1
print("겹치는 비율: {:.2f}".format(count/len(h_clustering.labels_)))
```

주어진 데이터에 이미 라벨링(정답 값이 존재)이 되어 있으면 군집화 진행을 할 필요가 없다.

하지만, 그렇지 않은 경우, 직접 라벨링이 어려울 때는 군집화를 통해 데이터가 어떻게 묶일 수 있는지 확인해볼 수 있다.

군집화는 어떤 features를 썼고, 몇 개의 집합을 설정했느냐에 따라서 결과가 달라지기 때문에 다양한 경우의 수를 고려해서 진행하거나 도메인 전문 지식을 활용해서 진행하면 된다.

# 3. Feature Engineering

차원 축소를 테스트 해보자.

```python
from sklearn.decomposition import MiniBatchSparsePCA as PCA

pca = PCA(n_components=2,
          batch_size=10,
          normalize_components=True,
          random_state=42)

X_pca = pca.fit_transform(X_scaled)
```

가장 높은 성능을 보여주었던 random forest 분류기를 사용한다.

```python
kfold = KFold(n_splits=5, shuffle=True)
scores = cross_val_score(RandomForestClassifier(n_estimators=100, random_state=123), 
                         X_pca, y_happiness['target_n'], cv=kfold)
print('교차검증 점수 \n{}'.format(scores))
print('교차검증 점수 평균 \n{:.3f}\n'.format(scores.mean()))
```

성능은 오히려 떨어졌다. 이는 차원 축소 결과가 각 그룹을 대표하는데 큰 효과가 없었음을 의미한다.
