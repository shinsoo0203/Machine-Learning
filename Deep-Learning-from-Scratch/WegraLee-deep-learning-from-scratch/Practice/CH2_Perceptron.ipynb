# Perceptron


- 다수의 신호를 입력으로 받아 하나의 신호를 출력  
- 뉴런=노드, 가중치, 임계값

### Logic circuit

```python
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1
```

```python
AND(0,0)
```

### Weight & Bias

```python
import numpy as np

x = np.array([0, 1])     #입력
w = np.array([0.5, 0.5]) #가중치
b = -0.7                 #편향
np.sum(w*x)+b
```

```python
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5]) #각 입력신호가 결과에 주는 영향력(중요도) 조절
    b = -0.7                 #뉴런이 얼마나 쉽게 활성화(결과=1)하느냐 조절
    tmp = np.sum(w*x)+b
    if tmp<=0:
        return 0
    else:
        return 1
```

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5]) #AND와는 가중치(w, b) 다르다
    b = 0.7
    tmp = np.sum(w*x)+b
    if tmp<=0:
        return 0
    else:
        return 1
```

```python
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(w*x)+b
    if tmp<=0:
        return 0
    else:
        return 1
```

- XOR : 좌표계에 배타적 논리합 퍼셉트론 표현 시, 선형으로 영역을 나눌 수 없음

# Multi-Layer Perceptron

- 층을 하나 더 쌓아 XOR 게이트 표현

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y
```

```python
XOR(1, 1)
```

- 단층 퍼셉트론으로 표현하지 못한 것 층을 늘려 구현 가능
- 2층 퍼셉트론, 비선형인 시그모이드 함수를 활성화 함수로 사용하여 임의의 함수 표현 가능
