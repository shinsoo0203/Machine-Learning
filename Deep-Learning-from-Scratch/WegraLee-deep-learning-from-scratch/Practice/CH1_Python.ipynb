# Numpy

```python
import numpy as np

x = np.array([1.0, 2.0, 3.0])
y = np.array([2.0, 4.0, 6.0])
print(x)
type(x)
x*y #element-wise product

A = np.array([[1, 2], [3, 4]])
print(A)
A.shape
A.dtype #행렬에 담긴 원소의 자료형
```

### Broadcast

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([10, 20])
A * B
```

```python
X = np.array([[51, 55], [14, 19], [0, 4]])
for row in X:
    print(row)
X = X.flatten()
print(X)
X > 15
X[X>15]
```

# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos")

plt.xlabel("x")
plt.ylabel("y")
plt.title('sin & cos')
plt.legend()
plt.show()
```
